# Triangle Classification and Visualization

## Import Required Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Import the classifyTriangle() Function

```python
from Triangle import classifyTriangle
```

## Define the Visualization Function

```python
def visualize_triangle(a, b, c):
    """
    Visualize a triangle given its side lengths.
    
    :param a: Length of side a
    :param b: Length of side b
    :param c: Length of side c
    """
    # Check for valid triangle
    if classifyTriangle(a, b, c) == 'Invalid':
        print("Cannot visualize an invalid triangle.")
        return

    # Calculate the coordinates of the triangle
    A = [0, 0]
    B = [a, 0]
    angle_C = np.arccos((a**2 + b**2 - c**2) / (2 * a * b))
    C = [b * np.cos(angle_C), b * np.sin(angle_C)]

    # Create a triangle plot
    triangle = plt.Polygon([A, B, C], fill=None, edgecolor='r')
    
    plt.gca().add_patch(triangle)
    plt.xlim(-1, max(a, b, c) + 1)
    plt.ylim(-1, max(a, b, c) + 1)
    plt.title(f'Triangle with sides: {a}, {b}, {c}')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid()
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

## Visualize Triangles

### In a New Cell, am calling the visualize_triangle() function with different sets of side lengths to visualize the triangles:

```python
visualize_triangle(3, 4, 5)  # Scalene triangle
```

```python
visualize_triangle(3, 3, 3)  # Equilateral triangle
```

```python
visualize_triangle(5, 5, 3)  # Isosceles triangle
```

```python
import pandas as pd

def run_tests():
    # Define the test cases as tuples of (a, b, c, expected)
    test_cases = [
        (3, 3, 3, 'Equilateral'),  # Test Equilateral triangle
        (5, 5, 3, 'Isosceles'),    # Test Isosceles triangle
        (3, 4, 5, 'Scalene'),      # Test Scalene triangle
        (1, 10, 12, 'Invalid'),    # Test Invalid triangle - sum of two sides less than the third
        (-1, 4, 5, 'Invalid'),     # Test Invalid triangle - negative side length
        (0, 4, 5, 'Invalid'),      # Test Invalid triangle - zero side length
        (1000000, 1000000, 1000000, 'Equilateral'),  # Test with large values
        (0.5, 0.5, 0.5, 'Equilateral'),  # Test with floating-point values
        (1, 1, 2, 'Invalid')       # Test boundary case where the sum equals the third side
    ]

    # Create lists to hold results for the DataFrame
    results = []

    # Run each test case
    for a, b, c, expected in test_cases:
        actual = classifyTriangle(a, b, c)
        status = "Pass" if actual == expected else "Fail"
        results.append((a, b, c, expected, actual, status))

    # Create a DataFrame from the results
    df = pd.DataFrame(results, columns=['Input a', 'Input b', 'Input c', 'Expected Result', 'Actual Result', 'Status'])

    # Display the DataFrame as a table
    print("Running Triangle Classification Tests:\n")
    display(df)  # Use display() to show DataFrame nicely in Jupyter
```

```python
# Call the function to run the tests and display results
run_tests()
```
